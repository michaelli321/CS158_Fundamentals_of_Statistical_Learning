```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
```

```python
N =[10**2, 10**3, 10**4]
P = np.arange(1, 1000)
outcomes = []
for n in N:
    res = []
    
    for p in P:
        res.append(integrate.quad(lambda d: n*p*d**p * (1-d**p)**(n-1), 0, 1)[0])
    
    outcomes.append(res)
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.plot(P, outcomes[0], label='N=10**2')
plt.plot(P, outcomes[1], label='N=10**3')
plt.plot(P, outcomes[2], label='N=10**4')
plt.xlabel('Dimension (p)')
plt.ylabel('Expectation of D')
plt.title('Expectation of D vs Dimension for Different N')
plt.legend()
plt.savefig('2a.png')
```
